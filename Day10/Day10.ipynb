```python
from pathlib import Path
from time import time

input_file = Path('../../AdventOfCode_inputs/AoC-2025-10-input.txt')

rinput = input_file.read_text()
```

```python
sinput = '''[.##.] (3) (1,3) (2) (2,3) (0,2) (0,1) {3,5,4,7}
[...#.] (0,2,3,4) (2,3) (0,4) (0,1,2) (1,2,3,4) {7,5,12,7,2}
[.###.#] (0,1,2,3,4) (0,3,4) (0,1,2,4,5) (1,2) {10,11,11,5,10,5}
'''
```

```python
ainput = rinput
```

```python
machines = [m.split(" ") for m in ainput.splitlines()]
machines = [ [[(1 if l=='#' else 0 ) for l in mach[0][1:-1]], 
              [list(map(int,b[1:-1].split(','))) for b in mach[1:-1] ], 
              list(map(int, mach[-1][1:-1].split(","))) ] for mach in machines ]
len(machines)
```

**Part One**

```python
from z3 import IntVector, Optimize
```

```python
startTime = time()
total = 0
for mach in machines:
    Ligs = IntVector('L', len(mach[0]))
    Buts = IntVector('b', len(mach[1]))
    o = Optimize()
    for b in Buts:
        o.add(b>=0)

    for i in range(len(mach[0])):
        o.add( sum(Buts[j] if i in mach[1][j] else 0 for j in range(len(mach[1])))==2*Ligs[i]+mach[0][i] )

    o.minimize(sum(Buts))
    o.check()
    for b in Buts:
        total += o.model().eval(b).as_long()
ans1=total
time()-startTime
```

**Part Two**

```python
startTime = time()

total = 0
for mach in machines:
    Buts = IntVector('b', len(mach[1]))
    o = Optimize()
    for b in Buts:
        o.add(b>=0)

    for i in range(len(mach[0])):
        o.add( sum(Buts[j] if i in mach[1][j] else 0 for j in range(len(mach[1])))==mach[2][i] )

    o.minimize(sum(Buts))
    o.check()
    for b in Buts:
        total += o.model().eval(b).as_long()
ans2=total

time()-startTime
```
