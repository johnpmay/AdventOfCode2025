**Part Two**

```python
fresh = [list(map(int,f.split('-'))) for f in ofresh.splitlines()]
while True :
    mfresh = list()
    skip = False
    for i in range(0,len(fresh)):
        if skip :
            skip = False
            continue
        if i+1 < len(fresh) and ( fresh[i][1]>=fresh[i+1][0] and fresh[i][1]<=fresh[i+1][1]  or fresh[i+1][1]>=fresh[i][0] and fresh[i+1][1]<=fresh[i][1]):
            mfresh.append([min(fresh[i][0],fresh[i+1][0]),max(fresh[i][1],fresh[i+1][1])])
            skip = True
        else:
            mfresh.append(fresh[i])
            skip = False
#        print(i, mfresh)
    if fresh != mfresh :
        fresh = mfresh
        fresh.sort()
    else:
        break
        
ans2 = sum(list(f[1]-f[0]+1 for f in fresh))
```
