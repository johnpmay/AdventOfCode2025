```python
from pathlib import Path
import functools

input_file = Path('../../AdventOfCode_inputs/AoC-2025-03-input.txt')

rinput = input_file.read_text()
```

```python
sinput = '''987654321111111
811111111111119
234234234234278
818181911112111
'''
```

```python
ainput = rinput
```

```python
batts = ainput.split()
```

**Part One**

```python
joltage = 0
for b in batts :
    bank = list(map(int,list(b))) # list of digits
    fd = max(bank)
    fdi = bank.index(fd)
    if fdi == len(bank)-1 :
        sd = fd
        fd = max(bank[:-1])
    else:
        sd = max(bank[fdi+1:])
    joltage += 10*fd+sd
    
ans1 = joltage
```

**Part Two**

```python
@functools.lru_cache(maxsize=None)
def _jolt(bank,l,u,n): # pass bank as a str b/c lists aren't cacheable
    if n == 0 :
        return 0 
    if u-l < n :
        return -1
    
    d = max(list(map(int,list(bank[l:u])))) # max digit
    if n == 1:
        return d
    i = (bank[l:u]).index(str(d))
    
    cands = []
    for p in range(0,n):
        # maximize d in pos p of n    
        lj = _jolt(bank,l,l+i,p)
        rj = _jolt(bank,l+i+1,u,n-p-1)
        if rj == -1 or lj == -1 :
            continue
        cands.append(lj*10**(n-p)+d*10**(n-p-1)+rj)
    if len(cands) == 0:
        return -1
    return max(cands)     
```

```python
joltage = 0
for b in batts:
    joltage += _jolt(b, 0, len(b), 12)
ans2 = joltage
```
