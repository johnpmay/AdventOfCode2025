```python
from pathlib import Path
import sympy, re, time

input_file = Path('../../AdventOfCode_inputs/AoC-2025-02-input.txt')

rinput = input_file.read_text()
```

```python
sinput = '''11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124
'''
```

```python
ainput = rinput
```

```python
ranges = [list(map(int, s.split('-'))) for s in ainput.splitlines()[0].split(',')]
```

**Part One**

```python
total = 0
for r in ranges :
    for i in range(r[0],r[1]+1):
        si = str(i)
        n = len(si)
        if n % 2 == 1 :
            continue
        if si[0:(n//2)] == si[(n//2):] :
            #print(i)
            total += i
ans1 = total
```

**Part Two**

```python
total = 0
startT = time.time()

for r in ranges :
    m = len(str(r[0]))
    divs = sympy.divisors(m)[0:-1]
    for i in range(r[0],r[1]+1):
        si = str(i)
        n = len(si)
        
        if n == 1 : # singletons are always valid
            continue
            
        if n != m : # update divisors if length changes
            m = n
            divs = sympy.divisors(m)[0:-1]
            
        for d in divs:
            # break si into d equal pieces of lenth l
            l = n//d
          
            if len( set(si[k:k+l] for k in range(0, n, l) ) ) == 1 :
                #print(l, i, r)
                total += i
                break
ans2 = total
print(time.time() - startT)
```
