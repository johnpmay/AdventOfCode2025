```python
from pathlib import Path

input_file = Path('../../AdventOfCode_inputs/AoC-2025-07-input.txt')

rinput = input_file.read_text()
```

```python
sinput = '''.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............
'''
```

```python
ainput = rinput
```

```python
grid = ainput.splitlines()
```

```python
start = grid[0].index('S')
start
```

**Part One**

```python
count = 0
beamloc = [start]
for i in range(1,len(grid)) :
    if grid[i].find('^') >= 0:
        splitloc = list( filter(lambda j: grid[i][j] == '^', range(0,len(grid[4]))) )
#        print(splitloc)
    else:
        continue
        
    newbl = set()
    for b in beamloc:
        if b in splitloc:
            newbl.add(b-1)
            newbl.add(b+1)
            count += 1
        else :
            newbl.add(b)
    beamloc = list(newbl)
#    print('b',beamloc)
ans1=count
```

**Part Two**

```python
count = 0
beamloc = {start: 1}
for i in range(1,len(grid)) :
    if grid[i].find('^') >= 0:
        splitloc = list( filter(lambda j: grid[i][j] == '^', range(0,len(grid[4]))) )
        
    newbl = { i:0 for i in range(0,len(grid[0]))}
    for b in beamloc.keys():
        if b in splitloc:
            newbl[b-1] += beamloc[b]
            newbl[b+1] += beamloc[b]
        else :
            newbl[b] += beamloc[b]
    beamloc = newbl
 #   print(beamloc)
ans2 = sum([i[1] for i in beamloc.items()])
```
