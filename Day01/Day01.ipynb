```python
sinput = '''L68
L30
R48
L5
R60
L55
L1
L99
R14
L82'''
```

```python
ainput = rinput
```

```python
turns = list( map(lambda x: (-1 if x[0]=='L' else 1)*int(x[1:]), ainput.split()) )
```

```python
dial = 50
count = 0

for x in turns:
    #print(dial,x)
    dial += x
    dial %= 100
    #print(dial)
    if dial == 0:
        count += 1
```

```python
ans1 = count
print(ans1)
```

```python
dial = 50
count = 0

for x in turns:
    ndial = dial + x
    ndial %= 100
    #print(dial,x,ndial,int(x/100))
    if (x > 0 and ndial < dial) or (x < 0 and dial != 0 and dial < ndial) or (ndial == 0):
        count += 1
    count += abs( int(x/100) )
    #print(count)
    dial = ndial
```

```python
ans2 = count
print(ans2)
```
