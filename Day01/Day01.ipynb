```python
from pathlib import Path

input_file = Path('../../AdventOfCode_inputs/AoC-2025-01-input.txt')

rinput = input_file.read_text()
```

```python
sinput = '''L68
L30
R48
L5
R60
L55
L1
L99
R14
L82'''
```

```python
ainput = rinput
```

```python
turns = [ (-1 if x[0]=='L' else 1)*int(x[1:]) for x in ainput.split() ]
```

**Part One**

```python
dial = 50
count = 0

for x in turns:
    dial += x
    dial %= 100
    if dial == 0:
        count += 1
```

```python
ans1 = count
print(ans1)
```

**Part Two**

```python
dial = 50
count = 0

for x in turns:
    ndial = dial + x
    ndial %= 100
    if (x > 0 and ndial < dial) or (x < 0 and dial != 0 and dial < ndial) or (ndial == 0):
        count += 1
    count += abs( int(x/100) )
    dial = ndial
```

```python
ans2 = count
print(ans2)
```
