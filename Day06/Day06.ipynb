```python
from pathlib import Path
import math
import numpy as np

input_file = Path('../../AdventOfCode_inputs/AoC-2025-06-input.txt')

rinput = input_file.read_text()
```

```python
sinput = '''123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
'''
```

```python
ainput = rinput
```

```python
grid = list(s.split() for s in ainput.splitlines())
```

```python
h = len(grid)
```

**Part One**

```python
total = 0
for c in range(0,len(grid[0])) :
    colop = grid[h-1][c]
    col = [ int(grid[r][c]) for r in range(0,h-1) ]
#    print(colop, col)
    if colop == '+':
        total += sum(col)
    else :
        total += math.prod(col)
total
```

**Part Two**

```python
tgrid = np.array(list(list(s) for s in ainput.splitlines())).transpose()
tgrid = tgrid.tolist()
stgrid = [ "".join(tgrid[i]) for i in range(0, len(tgrid))]
```

```python
total = 0
i = 0
while i < len(stgrid) :
    bop = stgrid[i][-1]
    bnums = [ int(stgrid[i][0:-1]) ]
    i += 1        
    while i < len(stgrid) and stgrid[i].strip() != '' :
        bnums.append( int(stgrid[i][0:-1]) )
        i += 1
        
    if bop == '+' :
        total += sum(bnums)
    else :
        total += math.prod(bnums)
        
    i += 1
    
total
```
