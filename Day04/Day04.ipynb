```python
from pathlib import Path

input_file = Path('../../AdventOfCode_inputs/AoC-2025-04-input.txt')

rinput = input_file.read_text()
```

```python
sinput = '''..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.
'''
```

```python
ainput = rinput
```

```python
ogrid = [list(s) for s in ainput.splitlines()] 
```

```python
l = len(ogrid[0])
h = len(ogrid)
l,h
```

```python
grid = [list("."*(l+2))] + [ ["."]+r+["."] for r in ogrid ] + [list("."*(l+2))] # padded grid for simpler indexing
```

**Part One**

```python
count = 0
for i in range(1,h+1) :
    for j in range(1,l+1) :
        if grid[i][j] == '.' :
            continue
        # there must be a better way to do 2-D slices
        c = 0
        c += grid[i-1:i+2][0][j-1:j+2].count('@')
        c += grid[i-1:i+2][1][j-1:j+2].count('@')
        c += grid[i-1:i+2][2][j-1:j+2].count('@')
        if c <= 4 :
            count += 1
ans1 = count
```

```python
ans1
```

**Part Two**

```python
grid = [list("."*(l+2))] + [ ["."]+r+["."] for r in ogrid ] + [list("."*(l+2))]

count = 0
ccount = 1
newgrid = grid
while ccount != 0 :
    ccount=0
    grid = newgrid # doing the easy CA thing with two grids
    for i in range(1,h+1) :
        for j in range(1,l+1) :
            if grid[i][j] == '.' :
                continue
            c = 0
            c += grid[i-1:i+2][0][j-1:j+2].count('@')
            c += grid[i-1:i+2][1][j-1:j+2].count('@')
            c += grid[i-1:i+2][2][j-1:j+2].count('@')
            if c <= 4 :
                ccount += 1
                newgrid[i][j]='.'
    count += ccount

ans2 = count
```
