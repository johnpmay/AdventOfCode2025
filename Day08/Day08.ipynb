```python
ainput = rinput
```

```python
boxes = [ list(map(int, s.split(','))) for s in ainput.splitlines() ]
n = len(boxes)
```

**Part One**

```python

D = np.array([[0]*n]*n, dtype=np.int64)
for i in range(n) :
    D[i,i] = 2**62
    for j in range(i+1,n) :
        D[i,j] = sum([(boxes[i][k]-boxes[j][k])**2 for k in range(3) ])
        D[j,i] = 2**62
        
circs = { i : i for i in range(n) }
rcircs = {i : set([i]) for i in range(n)}

if n == 20 :
    connects = 10
else:
    connects = 1000
    
for k in range(connects) :
    m = D.argmin()
    p1 = m // n
    p2 = m % n
    D[p1,p2] = 2**62
#    print(k, boxes[p1], boxes[p2])
    if p2 in rcircs[circs[p1]] :
        continue # boxes already connected
    oldcirc = rcircs[circs[p2]] # all boxes connected to p2
    rcircs[circs[p1]] = rcircs[circs[p1]].union(oldcirc)
    rcircs[circs[p2]] = set()
    
    for b in oldcirc :
        circs[b] = circs[p1] # each now in p1's circuit
#    print(circs)

```

```python
csizes = [len(x) for x in rcircs.values()]
csizes.sort()
math.prod(csizes[-3:])
```
